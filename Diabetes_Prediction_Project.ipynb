#Importing Libraries

```python
import pandas as pd
import numpy as np
import sklearn.datasets
```

#Importing dataset

```python
Diabetes_data=pd.read_csv('/content/Diabetes data.csv')
```

#Defining Target(Y) & Features(X)

```python
Y=Diabetes_data['diabetes']
```

```python
X=Diabetes_data[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
Y.shape
```

```python
X.shape
```

#Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=2529)
```

```python
X_train.head(10)
```

#Selecting Model For Prediction

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=500)
```

#Train Model(Fit Model)

```python
model.fit(X_train,Y_train)
```

#Prediction

```python
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

#Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred)
```

#So the Accuracy Score is 77%
